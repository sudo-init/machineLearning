## 1. 준비

패키지와 데이터를 불러옵니다.

```python
# 패키지 불러오기
library(data.table)
library(magrittr)
library(ggplot2)
library(ggcorrplot)

# 그래프 설정
theme_set(theme_minimal() + 
        theme(plot.title= element_text(size=24),
              axis.title= element_text(size=20),
              axis.text = element_text(size=14),
              legend.title=element_text(size=18),
              legend.text=element_text(size=14),
              strip.text.x = element_text(size = 16, color = 'white', face = 'bold'),
              strip.background = element_rect(color='black', fill='#002060', size=1))
             )
options(repr.plot.width = 12, repr.plot.height = 8, repr.plot.res = 150)

```

```python
# 데이터 불러오기
dt_train = rbindlist(lapply(list.files('train/', full.names=TRUE), fread))
dt_train[, head(.SD, 3)]

dt_valid = fread('validation/validation.csv')
dt_valid[, head(.SD, 3)]

dt_test1 = fread('test/test1.csv')[, phrase_:='test1'][]
dt_test2 = fread('test/test2.csv')[, phrase_:='test2'][]
dt_test3 = fread('test/test3.csv')[, phrase_:='test3'][]
dt_test = rbind(dt_test1, dt_test2, dt_test3)
```

<br>

## 2. train 데이터의 시간 분할

제공된 데이터는 1초 단위로 적재되어 있는데, 데이터 분할 과정 등에서 누락된 구간이 발생합니다.  
시각화할 때 쓸데없이 공백이 생기고, 새로운 구간이 시작될 때 변수 값들이 크게 바뀌는 상황들도 많아서 구역(phrase)를 미리 나눠놓고 살펴보려합니다.

```python
# 이전 시점과 1초 이상 차이가 나는 시점 찾기
dt_train[, fault_ := ifelse((timestamp - shift(timestamp, fill=0))==1, 0, 1)]

# 누적합으로 구역 일련번호 생성하기
dt_train[, phrase_:= as.factor(cumsum(fault_))]
```

<br>

이렇게 만든 구역번호와 facet 기능을 활용하면 그래프가 훨씬 깔끔해집니다.  
train 데이터는 1초 단위로 연속적인 6개 구간으로 분리됩니다.

```python
# 간단한 시계열 시각화(C01)
target_var = 'C01'

dt_train %>% 
  ggplot(aes_string('timestamp', target_var, group='2')) + 
  geom_path(alpha=0.3) + 
  facet_wrap(vars(phrase_), ncol=3, scale='free_x')+ 
  scale_x_datetime(date_breaks='1 day', date_labels='%m/%d') 
```

<br>

## 3. 상수 변수의 분리 

위의 그래프로 모든 변수의 그래프를 저장하고 살펴보니, 일부 변수는 파동의 형태가 아니라 상수였습니다.  
예를 들어 `C02`의 경우 모든 6개 구간에서 값이 0으로 동일합니다. 

```python
target_var = 'C02'

dt_train %>% 
  ggplot(aes_string('timestamp', target_var, group='2')) + 
  geom_path() + 
  facet_wrap(vars(phrase_), ncol=3, scale='free_x')+ 
  scale_x_datetime(date_breaks='1 day', date_labels='%m/%d') 
```

<br>

만약 validation이나 test 데이터의 `C02`가 0의 값을 가지는 상수가 아니라면 곧바로 *엇! 이상하다*라고 할 수 있을 텐데요, 주최측에서 그러지말라고 했지만 몰래 validation과 test 데이터에 대해서 `C02` 변수의 흐름을 살펴봤습니다. 첫번째 그래프에서 빨간점은 validation에서 `attack`값이 1인 지점인데요, 어차피 값이 0이네요.  

두번째 그래프는 세 테스트 데이터 시점의 `C02` 시계열 그래프인데, 세번째 test 데이터가 참 귀찮게 만듭니다. 순간적으로 1로 튀는 지점들이 있네요.  

공유된 내용들을 살펴보면 train 데이터로 분포나 추세나 파동을 확인하고, validation으로 임계점/기준점을 잡는 공통된 큰 흐름이 있는데요, 이런 상수 변수들 발라내서 따로 살펴볼 필요도 있어보입니다. 


```python
options(repr.plot.width = 12, repr.plot.height = 5, repr.plot.res = 150)

dt_valid %>% 
  ggplot(aes_string('timestamp', target_var, group='2')) +
  geom_path() +
  geom_point(aes(color=as.factor(attack)), data=dt_valid[attack==1]) +
  theme(legend.position = 'None')


dt_test %>% 
  ggplot(aes_string('timestamp', target_var, group='2')) +  
  geom_path() + 
  facet_wrap(vars(phrase_), ncol=3, scale='free_x')
```

<br> 

전체 변수 중에서 상수 변수를 발라내는 것은 너무나 쉽습니다. 분산이나 표준편차를 계산하면 되죠. 혹시라도 구간별로 상수값이 달라질 수 있어서 변수/구간별 표준편차를 계산하고, 표준편차가 항상 0인 변수들을 추려내서 `x_names0`로 저장했습니다. 20개나 됩니다.  
86개 중에서 이런 변수는 먼저 걸러내고 작업하는 것도 좋은 전략이 될 것 같네요.

```python
# 표준편차 -> 상수 변수 확인
x_names = paste0('C', formatC(1:86, width=2, flag='0'))
x_names0= x_names[which(dt_train[, lapply(.SD, sd, na.rm=T), by=.(phrase_), .SDcols=x_names][, sapply(.SD, max), .SDcols=x_names]==0)]
x_names0

x_names1= setdiff(x_names, x_names0)
# x_names1
```

<br>

## 4. 계단식 변화 변수 확인

상수 변수처럼 완전히 정적이진 않지만, 긴 시간 변하지 않다가 계단식으로 값이 바뀌는 변수들도 있습니다. 역시 찾아보겠습니다.  
계단이 1000개보다 적은 변수들만 선택했는데요, 12개가 있습니다.

```python
# 정적인 변수
dt_train[, paste0('phrase_', x_names1) := lapply(.SD, function(x) cumsum(fifelse(x==shift(x, fill=0), 0, 1))), .SDcols=x_names1]
x_names2 = x_names1[dt_train[, sapply(.SD, last), .SDcols=paste0('phrase_', x_names1)] <1000]
x_names2
```

```python
# 간단한 시계열 시각화(C07)
options(repr.plot.width = 12, repr.plot.height = 8, repr.plot.res = 150)
target_var = 'C61'

dt_train %>% 
  ggplot(aes_string('timestamp', target_var, group='2')) + 
  geom_path() + 
  facet_wrap(vars(phrase_), ncol=3, scale='free_x')+ 
  scale_x_datetime(date_breaks='1 day', date_labels='%m/%d') 
```

그 중 `C61` 변수를 그려봤습니다. 성곽 같네요.  
이런 계단식 변수는 무언가 **변화**를 의미하고, 보안에서 변화는 위험신호가 될 수 있습니다.  
그러지 말라고 했는데... 이번에도 validation과 test 데이터로도 그래프를 그려봅니다. 몇개 변수를 살펴봤지만 validation에서 빨간점으로 찍힌 `attack`이 1인 지점과 단차가 생기는 부분이 맞아 떨어지지는 않더라구요.  
얘네들도 딱히 중요해보이지 않습니다. 



```python
options(repr.plot.width = 12, repr.plot.height = 5, repr.plot.res = 150)

dt_valid %>% 
  ggplot(aes_string('timestamp', target_var, group='2')) +
  geom_path() +
  geom_point(aes(color=as.factor(attack)), data=dt_valid[attack==1]) +
  theme(legend.position = 'None')


dt_test %>% 
  ggplot(aes_string('timestamp', target_var, group='2')) +  
  geom_path() + 
  facet_wrap(vars(phrase_), ncol=3, scale='free_x')
```

<br>

## 5. 상관계수 

86개 변수들 중에서 20개는 상수 변수니까 제외하고, 계단 같은 변수 12개도 제외하면 54개가 남습니다.  
왠지 이 중에서 상관이 높은 변수들이 있지 않을까 생각했는데요, 생각보다 강한 상관관계가 많이 보입니다. 

```python
options(repr.plot.width = 12, repr.plot.height = 12, repr.plot.res = 150)

# 상관계수
x_names1_ = setdiff(x_names1, x_names2)
cor_x = dt_train[, ..x_names1_] %>% cor(use='pairwise.complete.obs')
p_cor = ggcorrplot(cor_x, type='upper', hc.order=T, outline.color='#FFFFFF11')
p_cor + scale_fill_distiller(palette='RdBu')
```

일반적으로 시그널들은 파동의 형태로 존재하고, 상관계수도 두 개의 파동이 얼마나 닮았는지를 따지는 것은 조금 무모하다고 생각했습니다. 그런데 아니더라구요.  
붉은 색으로 칠해진 삼각형이 몇개 보입니다. 별다른 변환을 안했는데도 강한 관계를 가진 변수 조합들이 나옵니다. 그리고 파란색 사각형도 몇개 보이는데요, 변수 조합끼리 서로 반대로 가는 경향을 가지고 있기도 하네요.  

남은 변수가 54개인데 이 그림을 보면, PCA 같은 방법으로 변수를 묶어보면 10개 내외의 성분으로 거의 다 설명이 가능할 것 같네요.  


변수 묶음을 만드는 것이 아니라 이 상관계수 행렬 자체를 이상치 탐지에도 활용할 수 있을 것 같습니다. 원래 강한 관계를 가지고 있는 두 변수의 관계가 갑자기 깨진다면, 특정 시간 구간에서의 상관계수의 패턴이 달라진 것으로 이상 징후를 포착할 수 있겠네요.
